```python
# gensim 
from gensim.models import KeyedVectors

#LOAD WORD2VEC MODEL
WORD2VEC_MODEL="/mnt/volume/GoogleNews-vectors-negative300.bin"

#The vectors can also be instantiated from an existing file on disk in the original Googleâ€™s word2vec format as a KeyedVectors instance
word2vec = KeyedVectors.load_word2vec_format(WORD2VEC_MODEL, binary=True) #float32
```

```python
word2vec['paris']
```

```python
word2vec.most_similar(positive=['woman', 'king'], negative=['man'])
```

```python
word2vec.doesnt_match("breakfast cereal dinner lunch".split())
```
